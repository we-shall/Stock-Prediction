# Check whether the news was missed while filtering
This process adds another level of check on the filtration of the news that whether the news that was of a NSE (National Stock Exchange) listed company was missed or not.

For this, we take our input as the file which we got after extraction of organization names from Stanford NER.

We then check this news from the list of NSE companies. If the news was of the NSE listed company, we append it to the LeftOutNews file. The fewer the number of entries in the LeftOutNews file, the better our efficiency.

```python
#Check if organization name of remaining news is present in NSE List
import csv

remaining = open(r'test6.csv','r',encoding='ISO-8859-1')
news = csv.reader(remaining)
next(news, None)
new=open(r'LeftoutNews.csv','a',newline='',encoding='ISO-8859-1')
writ=csv.writer(new)
writ.writerow(['Title','COMPANYNAME'])
for n in news:
    testnews = n[0]
    print('\n'+testnews)
    testorg1= str(n[1])
    print(testorg1)
    testorg=testorg1[:-1]
    #print('TEST ORG- ',testorg)
    list1=[]
    list1 = testorg.split(" ")
    list1.append("ZZZZZZZ")
    #print('Split List:', list1)
    org =[]
    i =0
    temp=''
    word = list1[0]
    while(i<len(list1)-1):
        temp = word + ' ' + list1[i+1]
        #print('temp = ', temp)
        if(temp in testnews ):
            word = temp

        else:
            org += [word]
            word = list1[i+1]
        i+=1
    print('Org : ',org)
    for j in org:
        nselist = open(r'NseData.csv','r',encoding='ISO-8859-1')
        rea=csv.reader(nselist)
        next(rea,None)
        for nse in rea:
            if j.upper()==nse[5] or j.upper()==nse[6] or j.upper()==nse[7]:
                writ.writerow([testnews,j])
remaining.close()
nselist.close()
```
